```python
import requests
import json
import pandas as pd
import xlwings
import xlwt

#先爬一条数据到文件
url="http://datacenter.eastmoney.com/api/data/get?&type=RPTA_WEB_JY_HQ&sty=ALL&filter=(date%3E%272012-05-06%27)&st=date&sr=-1&token=894050c76af8597a853f5b408b759f5d&p=1&ps=5000&source=WEB&_=1603599014837"
headers={"User_Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36",
    }
r=requests.get(url)
data=r.text
html_str=r.content.decode()
    # str = html_str[42:]
    # str=str[:-8]
```

```python
print(html_str)
```

```python
dict_json=json.loads(html_str)
close=[]
date=[]
qy=[]
cy=[]
for i in range(0,len(dict_json["result"]["data"])):
    data1=dict_json["result"]["data"][i]
    _close=data1["close"]
    _date=data1["date"]
    _qy=data1["qy"]
    _cy=data1["cy"]
    close.append(_close)
    date.append(_date)
    qy.append(_qy)
    cy.append(_cy)
dataTable={"date":date,"NYMEX原油":close,"汽油供应价格":qy,"柴油供应价格":cy}
df=pd.DataFrame(dataTable,columns=["date","NYMEX原油","汽油供应价格","柴油供应价格"])
#写文件，写入Excel或csv，看个人需求
writer = pd.ExcelWriter('D://桌面//数据集//oilprice_test.xls')
df.to_excel(excel_writer=writer, header=True,index=False) 
writer.save()
writer.close() 
```
