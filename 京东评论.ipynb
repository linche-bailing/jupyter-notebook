```python
#京东评论
import requests
from bs4 import BeautifulSoup
import bs4
import time
import json
h= {"User-Agent": "Mozilla/5.0",
    'referer': 'https://item.jd.com/7611478.html'}
url='https://club.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98&productId=7611478&score=0&sortType=5&page=1&pageSize=10&isShadowSku=0&rid=0&fold=1'

r=requests.get(url,headers=h, timeout=30)
r.raise_for_status()  #产生异常信息
r.encoding=r.apparent_encoding
rdata=r.text
ndata=rdata[20:-2]
datas = json.loads(ndata)
comments = datas['comments']
for comment in comments:
    content = comment['content']
    creationTime = comment['creationTime']
    days = comment['days']
    print( content, creationTime, days)
```
