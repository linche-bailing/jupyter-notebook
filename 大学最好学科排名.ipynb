```python
import requests
import bs4
from bs4 import BeautifulSoup
import csv
url="https://www.shanghairanking.cn/rankings/bcsr/2020/0101"
headers={"User-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36"}
r=requests.get(url,headers=headers)
r.encoding=r.apparent_encoding
soup=BeautifulSoup(r.text,"html.parser")
ulist=[]
for tr in soup.find('tbody').children:
    if isinstance(tr,bs4.element.Tag):
        tds = tr('td')
        ulist.append([tds[0].text.strip(),tds[1]('span')[0].text.strip(),tds[2].text.strip(),tds[3]('a')[0].text.strip(),tds[4].text.strip()])       
with open("F:/jupyter notebook/数据保存/大学最好学科排名.csv",'w',newline='',encoding='gbk') as csvf:
    csv1=csv.writer(csvf)
    csv1.writerow(("2020排名",'2019排名','排名层次','大学名称','分数'))
    for u in ulist:
        csv1.writerow((u[0],u[1],u[2],u[3],u[4]))
```
